# Concate & Merge

```python
# Concate operation
import pandas as pd
import numpy as np
```

```python
df1 =pd.DataFrame({"A":["A0","A1","A2","A3"],"B":["B0","B1","B2","B3"],"C":["C0","C1","C2","C3"],"D":["D0","D1","D2","D3"]},index=[0,1,2,3])
df1
```

```python
df2 =pd.DataFrame({"A":["A4","A5","A6","A7"],"B":["B4","B5","B6","B7"],"C":["C4","C5","C6","C7"],"D":["D4","D5","D6","D7"]},index=[0,1,2,3])
df2
```

```python
df3 =pd.DataFrame({"A":["A8","A9","A10","A11"],"B":["B8","B9","B10","B11"],"C":["C8","C9","C10","C11"],"D":["D8","D9","D10","D11"]},index=[0,1,2,3])
df3
```

```python
pd.concat([df1,df2,df3],axis=0)
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
df4 =pd.DataFrame({"A":["A12","A13","A14","A15"],"B":["B12","B13","B14","B15"],"Z":["C12","C13","C14","C15"],"E":["D12","D13","D14","D15"]},index=[0,1,2,3])
df4
```

```python
pd.concat([df1,df2,df3,df4],axis=0)
```

```python
# Merging
left=pd.DataFrame({"key":["k0","k1","k2","k3"],
                  "A":["A0","A1","A2","A3"],
                  "B":["B0","B1","B2","B3"]})
left
```

```python
right=pd.DataFrame({"key":["k1","k2","k0","k4"],
                  "C":["C0","C1","C2","C3"],
                  "D":["D0","D1","D2","D3"]})
right
```

```python
pd.merge(left,right,on="key",how="inner") # inner means common values 
```

```python
pd.merge(left,right,on="key",how="outer")
```
