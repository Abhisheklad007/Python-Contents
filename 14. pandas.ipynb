```python
# Creat dataframe using dictionary
import pandas as pd
import numpy as np
```

```python
d = {"A":[1,2,3],"B":[4,5,6]}
pd.DataFrame(d)
```

```python
# Creating of dataframe using numpy

mat = np.random.randn(5,4)
pd.DataFrame(mat)
```

```python
gig = np.random.randn(5,4)
pd.DataFrame(gig,index="a,b,c,d,e".split(","),columns="w x y z".split())
```

```python
pd.DataFrame(gig,index=["A","B","C","D","E"],columns=["W","X","Y","Z"])
```

```python
my=pd.DataFrame(np.arange(1,21).reshape(5,4),index="a b c d e".split(),columns="W X Y Z".split())
my
```
